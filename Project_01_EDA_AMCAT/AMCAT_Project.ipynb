##### **Q :** Of the above columns, few are supposed to be of date type. Identify and fix their data type.

```python
for col in ['DOJ', 'DOL', 'DOB']:
    print(data[col][0], data[col][49])
```

It is clear that :

* in columns DOJ and DOB, the date is given in the format "m/d/Y H:M"
* in column DOL, either date is present in the format "m/d/Y H:M" or the value 'present' is seen
* we know that the data was collected in the year 2015, so we can replace it with some valid date in the year 2015, say "3/31/15 0:00"

```python
# fixing datatype of DOJ column
data['DOJ'] = data['DOJ'].str.replace(" 0:00", "", regex=False).str.replace("/", "-", regex=False)
data['DOJ'] = pd.to_datetime(data['DOJ'], format = "mixed")
```

```python
# fixing datatype of DOB column
data['DOB'] = data['DOB'].str.replace(" 0:00", "", regex=False).str.replace("/", "-", regex=False)
data['DOB'] = pd.to_datetime(data['DOB'], format = "mixed")
```

```python
# fixing datatype of DOL column
data['DOL'] = data['DOL'].str.replace("present", "3-31-15").str.replace(" 0:00", "", regex=False).str.replace("/", "-", regex=False)
data['DOL'] = pd.to_datetime(data['DOL'], format = "mixed")
```

##### **Q :** Set the datatypes of other object columns

```python
# rest of the object columns contain categorical variables, so fixing their datatypes suitably
for col in data.columns:
    if data[col].dtype == "object":
        data[col] = data[col].astype('str')
```

##### **Q :** Visualize the distribution of the target variable 'Salary' viz. of continuous numeric type

```python
# Create a figure with 2 rows and 2 columns
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(6,6))

# Plot KDE on the first subplot
data['Salary'].plot(kind='kde', ax=axes[0, 0], grid=True)
ax=axes[0, 0].set_xlabel('Salary of the graduate')
ax=axes[0,0].set_title('Kernel Density Estimation', color='red')

# Plot histogram on the second subplot
data['Salary'].plot(kind='hist', ax=axes[0, 1], grid=True)
ax=axes[0, 1].set_xlabel('Salary of the graduate')
ax=axes[0,1].set_title('Histogram', color='green')

# Plot box plot on the third subplot
data['Salary'].plot(kind='box', ax=axes[1, 0], grid=True)
ax=axes[1,0].set_title('Box Plot', color='orange')

# Remove the empty subplot
fig.delaxes(axes[1, 1])

# Set a common title for the figure
fig.suptitle('Salary Distribution Analysis', fontsize=14, color='purple')

plt.tight_layout()
plt.show()
```

* from the boxplot, it is clear that some extreme high values are present. These outliers shall be removed for better analysis.
* The histogram also indicates the presnce of some record with extreme high salary like 3M, 4M etc.

```python
# removing outliers
data = data[data['Salary']<=1000000]
```

**After removing outliers in Salary**

```python
# Create a figure with 2 rows and 2 columns
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8,8))

# first subplot
data['Salary'].plot(kind='kde', ax=axes[0, 0], grid=True)
ax=axes[0, 0].set_xlabel('Salary of the graduate')
ax=axes[0,0].set_title('Kernel Density Estimation', color='red')
ax=axes[0,0].axvline(325000, color='lime', linestyle='dotted')
ax=axes[0,0].axvline(200000, color='lime', linestyle='dotted')

# second subplot
data['Salary'].plot(kind='hist', ax=axes[0, 1], grid=True)
ax=axes[0, 1].set_xlabel('Salary of the graduate')
ax=axes[0,1].set_title('Histogram', color='green')

# third subplot
data['Salary'].plot(kind='box', ax=axes[1, 0], grid=True)
ax=axes[1,0].set_title('Box Plot', color='orange')
ax=axes[1,0].axhline(300000, color='lime', linestyle='dotted')

# Remove the empty subplot
fig.delaxes(axes[1, 1])

# Set a common title for the figure
fig.suptitle('Salary Distribution Analysis', fontsize=14, color='purple')

plt.tight_layout()
plt.show()
```

**Infrences** 

**From KDE Plot :**

1. The distribution is almost normal with a righward skew
2. The distribution is multimodal ie., has 2 peaks, 1 at Rs.200K and 1 at Rs.325K, indicating the presence of many graduates with these salaries in comparison with other values
3. the distribution is leptokurtic ie., it has sharp peaks

**From Histogram :**

1. Almost 3000 graduates earn an income between Rs.200K - Rs.400K.
2. About 500 graduates have a comparitively lower income of Rs.50K - Rs.100K
3. Less than 100 graduates have salary over 600K

**From the Boxplot :**

1. The median salary is Rs.300K
2. The median line(green) is more close to the upper end of the box, indicating a positive(right) skew. ie., majority of the graduates have salary close to 300K, but very few have really high salary
3. The IQR (bounded by the box) is Rs.175K-375K; ie., this is the salary range of majority of the graduates
4. The salary can be as low as Rs.35K as indicated by the lowest horizontal line
5. The salary can be as high as Rs.600K as indicated by the topmost horizontal line
6. The upper whisker is more longer than the lower whisker indicating the positive skew
